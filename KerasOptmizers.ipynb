```python
A = []
Y = []
for _ in range(250):
    y = random.sample(range(0, 256), 2)
    a = x[y[0]]
    b = x[y[1]]
    result = list(s^t for s,t in zip(a,b))
    A.append(a)
    A.append(b)
    Y.append(result)
```

```python
len(A)
```

```python
X = []
for i in range(0,len(A),2):
    l = []
    for j in list(A[i]):
        l.append(j)
    for j in list(A[i+1]):
        l.append(j)
    X.append(l)
```

```python
input_layer = input_data(shape=[None, 16])
hidden_layer1 = fully_connected(input_layer , 24, activation='sigmoid')
hidden_layer2 = fully_connected(hidden_layer1 , 24, activation='sigmoid')
hidden_layer3 = fully_connected(hidden_layer2, 16, activation='sigmoid')
output_layer = fully_connected(hidden_layer3, 8, activation='sigmoid')
```

```python
regression1 = regression(output_layer , optimizer='Adam', loss='binary_crossentropy', learning_rate=5)
model = DNN(regression1)
```

```python
import tflearn
class EarlyStoppingCallback(tflearn.callbacks.Callback):
    def __init__(self, val_acc_thresh):
        self.val_acc_thresh = val_acc_thresh

    def on_epoch_end(self, training_state):
        if training_state.val_acc is None: return
        if training_state.val_acc > self.val_acc_thresh:
            raise StopIteration

early_stopping_cb = EarlyStoppingCallback(val_acc_thresh=0.95)
model.fit(X,Y, n_epoch=8000, show_metric=True, snapshot_epoch=False, callbacks=early_stopping_cb)
```
